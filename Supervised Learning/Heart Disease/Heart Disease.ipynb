```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('heart_disease_data.csv')
```

```python
X = df.drop('target', axis = 1)
Y = df['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
rf = RandomForestClassifier()
```

```python
knn = KNeighborsClassifier()
```

```python
lr = LinearRegression()
```

```python
rf.fit(X_train, Y_train)
```

```python
train_prediction = rf.predict(X_train)
```

```python
accuracy_train = accuracy_score(train_prediction, Y_train)
```

```python
accuracy_train
```

```python
test_prediction = rf.predict(X_test)
```

```python
accuracy_test = accuracy_score(test_prediction, Y_test)
```

```python
accuracy_test
```

```python
knn.fit(X_train, Y_train)
```

```python
train_new_prediction = knn.predict(X_train)
```

```python
accuracy_train_new = accuracy_score(train_new_prediction, Y_train)
```

```python
accuracy_train_new
```

```python
test_new_prediction = knn.predict(X_test)
```

```python
accuracy_test_new = accuracy_score(test_new_prediction, Y_test)
```

```python
accuracy_test_new
```

```python
lr.fit(X_train, Y_train)
```

```python
train_lr = lr.predict(X_train)
```

```python
train_lr_binary = (train_lr >= 0.5).astype(int)
accuracy_lr = accuracy_score(train_lr_binary, Y_train)
```
