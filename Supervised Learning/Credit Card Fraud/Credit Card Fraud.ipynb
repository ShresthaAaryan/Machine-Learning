```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
 df.shape
```

```python
df.isnull().sum()
```

```python
valid = df[df.Class == 0]
```

```python
fraud = df[df.Class == 1]
```

```python
valid.shape
```

```python
fraud.shape
```

```python
valid_sample = valid.sample(n=492)
```

```python
valid_sample.shape
```

```python
new_df = pd.concat([valid_sample, fraud], axis=0)
```

```python
new_df['Class'].value_counts()
```

```python
new_df.shape
```

```python
X = new_df.drop('Class', axis = 1)
X_new = new_df.drop('Time', axis = 1)
Y = new_df['Class']
```

```python
X.head()
```

```python
X_new.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 42) 
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
train_prediction = model.predict(X_train)
```

```python
accuracy_train = accuracy_score(train_prediction, Y_train)
```

```python
accuracy_train
```

```python
test_prediction = model.predict(X_test)
```

```python
accuracy_test = accuracy_score(test_prediction, Y_test)
```

```python
accuracy_test
```

```python
X_new_train, X_new_test, Y_train, Y_test = train_test_split(X_new, Y, test_size = 0.2, stratify = Y, random_state = 42) 
```

```python
scaler = StandardScaler()
X_new_train = scaler.fit_transform(X_new_train)
X_new_test = scaler.transform(X_new_test)
```
